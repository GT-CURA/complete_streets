```python
import json
import nbformat
```

```python
with open("0_sign_detection.ipynb") as f:
    meta = json.load(f)["metadata"]["kernelspec"]
print(meta)
```

```python
notebook_path = "0_sign_detection.ipynb"
kernel_name = "env_street_parking"
kernel_display_name = "Python (env_street_parking)"

# Load the notebook
nb = nbformat.read(notebook_path, as_version=nbformat.NO_CONVERT)

# Update the metadata
nb['metadata']['kernelspec'] = {
    "name": kernel_name,
    "display_name": kernel_display_name,
    "language": "python"
}

# Save it back
nbformat.write(nb, notebook_path)

print(f"✅ Kernel metadata updated to '{kernel_display_name}' in {notebook_path}")
```
